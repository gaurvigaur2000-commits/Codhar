Check whether the remainder after division is zero.

```python
a, b = map(int, input("Enter a and b: ").split())
print("True" if a%b==0 else "False")
    
```

Check if the average of 3 numbers is above 60.

```python
list1 = list(map(int, input("Enter 3 numbers: ").split()))
avg = sum(list1)/len(list1)
print(f"Average: {avg:.2f}")
print("True" if avg>60 else "False")
```

Check if a number is greater than its reverse.

```python
num = int(input("Enter number: "))
rev = int(str(num)[::-1])
print("Yes" if num>rev else "No")
```

```python
num = int(input("Enter number: "))
rev, num2 = 0, num
while num:
    rev=rev*10+num%10
    num//=10
print("Yes" if num2>rev else "No")
```
